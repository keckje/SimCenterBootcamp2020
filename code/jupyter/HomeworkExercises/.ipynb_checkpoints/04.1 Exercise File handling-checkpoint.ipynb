## Problem 1: 

Open and read the file '../SelfStudyResources/ExampleData.txt'
.  This file is a plain text file providing data as comma-separated list.  An example of this file looks like this:
~~~
1.0,0.0,
1.0043021179718516,0.06331229237152888,
1.0045749747914792,0.12716436093274708,
1.0007678438038003,0.1912984977208162,
...
~~~
Each line represents the X and Y coordinate on one point on a polygon.

**Your tasks**: 
1. read this file making sure to implement proper exception handling.
2. plot the polygon.  Make sure to preserve correct proportions and angles by setting equal axis scaling.

```python
try:
    f = open('../SelfStudyResources/ExampleData.txt')
except IOError:
    print('unable to open file')
#     raise
except FileNotFoundError:
    print('name of file not correct')
    
x = []
y = []
cnt = 0
for line in f: # memory used is only one line
#     print(cnt,': ',line, end ='')
    x_i = line.split(',')[0]
    x.append(x_i)
    y_i = line.split(',')[1]
    y.append(y_i)
    cnt +=1
x = list(map(float, x)); y = list(map(float,y)) # convert list of str to float
plt.plot(x,y)
plt.axis('equal')
plt.show() # needed to plot if run as a script outside of jupyter
```
